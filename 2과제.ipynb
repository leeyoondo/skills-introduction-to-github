```python
scores = [80, 65, 43, 78, 88, 100, 44, 55, 75]
student_index = 0

for score in scores:
    student_index += 1
    if score >= 70:
        print(f'{student_index}번 학생 축하합니다!! 광주 소프트웨어마이스터고등학교에 최종 합격하셨습니다.')
    else:
        print(f'{student_index}번 학생 아쉽지만 다음 기회에 다시 지원해 주시길 바랍니다.')
```

```python
scores = [66,54,89,92,98,100,33,45,97,99,67,87]

high_scores_sum = 0
high_scores_count = 0
low_scores_sum = 0
low_scores_count = 0

for score in scores:
    if score >= 90:
        high_scores_sum += score
        high_scores_count += 1
    elif score < 60:
        low_scores_sum += score
        low_scores_count += 1

high_scores_avg = high_scores_sum / high_scores_count if high_scores_count > 0 else 0
low_scores_avg = low_scores_sum / low_scores_count if low_scores_count > 0 else 0

print(f'90점 이상 학생들의 평균 점수 : {high_scores_avg:.1f}, 90점 이상 학생 수 : {high_scores_count}명')
print(f'60점 미만 학생들의 평균 점수 : {low_scores_avg:.1f}, 60점 미만 학생 수 : {low_scores_count}명')
```
